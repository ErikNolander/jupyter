```python
import pandas as pd

# Läs in CSV-filen
df = pd.read_csv("assignment_3_dataset.csv")

# Datastädning
df.replace(["?", "??"], pd.NA, inplace=True)
df.drop(columns=["stmtL"], inplace=True)
df = df.applymap(lambda s: s.lower() if isinstance(s, str) else s)
df = df.applymap(lambda s: s.strip() if isinstance(s, str) else s)

# Konvertera alla numeriska kolumner till rätt typ
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="ignore")

# --- 🔹 Fyll saknade värden i numeriska kolumner med median ---
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns
for col in numeric_cols:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)
# --------------------------------------------------------------

# --- 🔹 Fyll saknade värden i kategoriska kolumner med modus ---
if "whours" in df.columns:
    median_whours = df["whours"].median()
    df.loc[df["whours"] > 10000, "whours"] = median_whours
# --------------------------------------------------------------

# Spara resultatet
df.to_csv("assignment_3_dataset_clean.csv", index=False)

# Visa DataFrame i Jupyter
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
df
```
